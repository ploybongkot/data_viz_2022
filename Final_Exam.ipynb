## การตรวจโรค(ov)

```python
alldb =  pd.read_csv(os.path.join(path,'alldb.csv')) ## ทำการนำเข้าข้อมูล การตรวจโรค(ov) โดยการเรียกใช้ข้อมูลประเภท .csv ใน drive ไฟล์ที่เรียกใช้คือ alldb.csv # โดยตั้งชื่ตารางว่า alldb
alldb.head() ## แสดงตาราง alldb
```

```python
alldb.shape     # ดูขนาด
```

```python
alldb = alldb.drop(columns=['Unnamed: 0','care','pid'])     # ลบ คอลัมน์ที่ไม่ต้องการออก
print(alldb.shape)
alldb.head()
```

```python
alldb.isnull().sum()    # ดูจำนวนค่าว่าง
```

```python
alldb.dtypes  # ดูชนิดข้อมูล
```

### เปลี่ยนข้อมูลให้เป็นไปตามข้อกำหนด

user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['user'].astype(str).str.isdigit()) & (alldb['user'].str.len() == 5)]   ## เลือกข้อมูลเฉพาะข้อมูลที่เป็นตัวเลข และ มีความยาว เท่ากับ 5
print(alldb.shape)
alldb.head()
```

```python
alldb.isnull().sum()  ## ดูค่าว่าง
```

RowKey เป็น String

```python
alldb['RowKey'] = alldb['RowKey'].astype('str') ## แปลง column RowKey ให้เป็น string
print(alldb.shape) ## ดูขนาดของข้อมูล
alldb.head() 
```

date เป็น YYYY-MM-DD

```python
import datetime ## ทำการ import package datetime 
alldb = pd.DataFrame(alldb) 
alldb['date'] = pd.to_datetime(alldb['date'], format = "%Y-%m-%d") 
print(alldb.shape)
alldb.head()
```

```python
alldb['date'] = alldb[['date']].fillna(method="ffill")   # เติมค่าว่างของวันที่ด้วย ค่าก่อนหน้า หรือค่าข้างบน
print(alldb.shape)
alldb.head()
```

phone เป็น String

```python
alldb['phone'] = alldb['phone'].astype('str') # แปลง column phone ให้เป็น string
print(alldb.shape)
alldb.head()
```

site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['site'].astype(str).str.isdigit()) & (alldb['site'].str.len() == 5)]  ## .str.isdigit ใช้ตรวจสอบว่าอักขระทั้งหมดเป็นตัวเลขหรือไม่
print(alldb.shape) 
alldb.head()
```

```python
alldb.isnull().sum()      ## ดูค่าว่าง
```

time เป็น hh-mm-ss

```python
alldb['time'] = alldb[['time']].fillna(method="ffill")      # เติมค่าว่างของเวลาด้วย ค่าก่อนหน้า หรือค่าข้างบน
print(alldb.shape)
alldb.head()
```

```python
import datetime ## ทำการ import package datetime 
alldb = pd.DataFrame(alldb) 
alldb['time'] = pd.to_datetime(alldb['time'],format = '%H:%M:%S').dt.time
print(alldb.shape)
alldb.head()
```

urine มีค่า String

```python
alldb['urine'] = alldb[['urine']].fillna(method="ffill")      # เติมค่าว่างของ urine(ปัสสาวะ) ด้วย ค่าก่อนหน้า หรือค่าข้างบน
print(alldb.shape)
alldb.head()
```

```python
alldb['urine'] = alldb['urine'].astype('str') ## แปลง column urine ให้เป็น string
print(alldb.shape)
alldb.head()
```

ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
alldb = alldb.dropna(subset=['ov']) ## ลบข้อมูลแถวที่ column ov ที่มีค่า missing
print(alldb.shape)
alldb.head()
```

```python
set(alldb['ov']) # ดู set ข้อมูลของ column ov
```

```python
alldb['ov'].replace({'ติดเชื้อ +2 ':'1'}, inplace=True) # แทนค่าติดเชื่อ + 2 ด้วย 1
```

```python
set(alldb['ov']) # ดู set ข้อมูลของ column ov
```

```python
alldb['ov'] = alldb['ov'].astype('int') ## แปลง column ov ให้เป็น int
print(alldb.shape)
alldb.head()
```

```python
alldb.isnull().sum() # ดูจำนวน missing ของแต่ละ column
```

```python
alldb = alldb[(alldb['PartitionKey'].astype(str).str.isdigit()) & (alldb['PartitionKey'].str.len() == 5)]  ## เลือกข้อมูลเฉพาะข้อมูลที่เป็นตัวเลข และ มีความยาว เท่ากับ 5
print(alldb.shape)
alldb.head()
```
