1.   คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด

````
this_data = data_covid[['sex','age','province_of_onset']]
````

````
this_data
````

````
#1. คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ของข้อมูลทั้งหมด
male = this_data[this_data['sex']=='ชาย']
male['age']
````

````
# หาค่าเฉลี่ยของอายุได้โดย
sum = 0     # กำหนดให้ sum = 0 เนื่องจากถ้าค่าตัวที่ 1 มาบวกจะได้ค่าออกมาเท่ากับตัวที่ 1
N = 0   # กำหนดให้ N = 0 เป็นตัวหาร โดยจะทำการวนลูป # N เป็นตัวแปรที่จะนับว่ามีคนทั้งหมดกี่คน 
for a in male['age']:   #เริ่มวนลูปโดยให้ a แทนค่าแต่ละค่าในคอมลัมน์ age ของตาราง male
    
    if a > 0:   # ถ้า a มากกว่า 0 แสดงว่าเป็นอายุปกติที่ไม่ใช่ข้อมูลที่ผิดพลาด
        sum += a    # sum = sum + a # ถ้า a > 0 ก็จะนำมารวมเข้าไปใน sum เลื่อยๆในทุกๆรอบที่ a มากกว่า 0
        N += 1  # คำนวนจำนวนโดยที่มันจะ +1 เข้ามาเลื่อยๆ
        #   sum คือค่าผลรวมทั้งหมด  N คือค่าจำนวนคนทั้งหมด
print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}')   # sum/N = หาอายุเฉลี่ยของคนทั้งหมด
````

````
#2.คำนวณ อายุเฉลี่ย ของ*ผู้ชาย *ของผู้ป่วยในจังหวัดขอนแก่น
male_KK = male[male['province_of_onset']=='ขอนแก่น']
sum = 0     # กำหนดให้ sum = 0 เนื่องจากถ้าค่าตัวที่ 1 มาบวกจะได้ค่าออกมาเท่ากับตัวที่ 1
N = 0   # กำหนดให้ N = 0 เป็นตัวหาร โดยจะทำการวนลูป # N เป็นตัวแปรที่จะนับว่ามีคนทั้งหมดกี่คน 
for a in male_KK['age']:  #เริ่มวนลูปโดยให้ a แทนค่าแต่ละค่าในคอมลัมน์ age ของตาราง male_KhonKaen
    
    if a > 0:   # ถ้า a มากกว่า 0 แสดงว่าเป็นอายุปกติที่ไม่ใช่ข้อมูลที่ผิดพลาด
        sum += a    # sum = sum + a  # sum = sum + a # ถ้า a > 0 ก็จะนำมารวมเข้าไปใน sum เลื่อยๆในทุกๆรอบที่ a มากกว่า 0
        N += 1  # คำนวนจำนวนโดยที่มันจะ +1 เข้ามาเลื่อยๆ
        # sum คือค่าผลรวมทั้งหมด  N คือค่าจำนวนคนทั้งหมด
print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย ในจังหวัดขอนแก่น {sum/N}')     # sum/N = หาอายุเฉลี่ยของคนทั้งหมด
````
